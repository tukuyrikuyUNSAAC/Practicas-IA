Importancia de la Preparación de Datos<br>
1. Los datos reales pueden ser impuros, pueden conducir
a la extracción de patrones/reglas poco útiles.<br>
Esto se puede deber a:<br>
Datos Incompletos: falta de valores de atributos, …<br>
Datos con Ruido<br>
Datos inconsistentes (incluyendo discrepancias)<br>

2. La preparación de datos puede generar un conjunto de
datos más pequeño que el original, lo cual puede
mejorar la eficiencia del proceso de Minería de Datos.<br>
Esta actuación incluye:<br>
Selección relevante de datos: eliminando registros
duplicados, eliminando anomalías, …<br>
Reduccion de Datos: Selección de instancias, selección de características,
muestreo o selección de instancias, discretización.<br>

3. La preparación de datos genera “datos de calidad”, los
cuales pueden conducir a patrones/reglas de calidad.<br>
Por ejemplo, se puede:<br>
Recuperar información incompleta.<br>
Eliminar outliers<br>
Resolver conflictos, …<br>

Filtrar las columnas (variables) cuyo porcentaje de filas (registros) únicos sea cercano a 1 porque lo más probable es que sean identificadores (IDs), hecho que normalmente no aporta información respecto a un objetivo como puede ser predecir los valores de una columna (variable) y por ende son considerados como ruido:

Determinar que columnas (variables) son del tipo númerico y cuales son categóricas para poder usar esta información más adelante:



```
# -- Crear listas para determinar columnas numéricas y categóricas
numeric_columns = list()
categorical_columns = list()
# -- Crear diccionarios para preparar datos que permitan la posterior
#    transformación de datos categóricos
dictionary_of_columns_with_index_to_categorical = dict()
dictionary_of_columns_with_categorical_to_index = dict()
# -- Procesar cada columna del dataset
for column in dataset:
  #Determinar si la columna (variable) es numérica o no
  if pd.api.types.is_numeric_dtype(dataset[column]):
    numeric_columns.append(column)
  else:
    #Modificar el tipo de dato de la columna (variable) mediante "astype"
    dataset[column] = dataset[column].astype("category")

    #Verificar si el tipo de dato de la columna (variable) fue transformado
    #a categórico correctamente
    if not pd.api.types.is_categorical_dtype(dataset[column]):
      raise Exception("La columna {} no se transformó correctamente a categórica".format(column))

    dictionary_of_columns_with_index_to_categorical[column] = dict()
    dictionary_of_columns_with_categorical_to_index[column] = dict()

    #Indexar los valores (categorías), sin tomar en consideración los nulos,
    #de la columna (variable) y guardamos esa información en los diccionarios
    for index, category in enumerate(dataset[column].cat.categories):
      dictionary_of_columns_with_index_to_categorical[column][index] = category
      dictionary_of_columns_with_categorical_to_index[column][category] = index

    categorical_columns.append(column)
```

##***2.5. Reducción de dimensionalidad***

La técnica de reducción de dimensionalidad se define como una forma de convertir un conjunto de datos de dimensiones elevadas en un conjunto de datos de dimensiones menores, asegurando que la información que proporciona es similar en ambos casos.<br>
Esta técnica se emplea a menudo en el aprendizaje automático para obtener un modelo predictivo más ajustado mientras se resuelven los problemas de regresión y clasificación que presentan los algoritmos.

###***2.5.1. Aplicación de Análisis de componentes principales (PCA)***

El análisis de componentes principales (PCA) es un procedimiento estadístico que utiliza una transformación ortogonal que convierte un conjunto de variables correlacionadas en un conjunto de variables no correlacionadas. PCA es la herramienta más utilizada en el análisis exploratorio de datos y en el aprendizaje automático para modelos predictivos. Además, PCA es una técnica estadística no supervisada utilizada para examinar las interrelaciones entre un conjunto de variables.
