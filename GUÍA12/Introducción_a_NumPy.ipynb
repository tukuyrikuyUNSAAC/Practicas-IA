```python
import numpy as np
```

## Crear Arreglos de diferentes dimensiones

```python
value = 10 #rank 0
value_as_array = np.array(value)

value_as_array
```

```python
value = [21, 83, 42, 35] #rank 1
value_as_array = np.array(value)

value_as_array
```

```python
value = [[21, 83, 42],
         [35, 12, 67]] #rank 2
value_as_array = np.array(value)

value_as_array
```

```python
value = [ [[21, 83, 42],
           [35, 12, 67]],
          [[56, 98, 24],
           [64, 78, 98]] ] #rank 3
value_as_array = np.array(value)

value_as_array
```

## Operaciones con Arreglos

```python
value_1 = [[21, 83],
           [42, 35]]
value_2 = [[95, 78],
           [65, 32]]

array_1 = np.array(value_1)
array_2 = np.array(value_2)
```

### Suma

```python
np.add(array_1, array_2)
```

```python
array_1 + array_2
```

### Resta

```python
np.subtract(array_1, array_2)
```

```python
array_1 - array_2
```

### Multiplicación de elementos

```python
np.multiply(array_1, array_2)
```

```python
array_1 * array_2
```

### Multiplicación de matrices

```python
#N,M   M,L   ->  N,L
np.matmul(array_1, array_2)
```

```python
array_1 @ array_2
```

## Obtener la transpuesta

```python
value = [ [[21, 83, 42],
           [35, 12, 67]],
          [[56, 98, 24],
           [64, 78, 98]] ]
value_as_array = np.array(value)
```

```python
np.transpose(value_as_array)
```

```python
value_as_array.T
```

## Obtener la inversa

```python
value = [[21, 83],
         [35, 12]]
value_as_array = np.array(value, dtype=np.float32)
```

```python
value_as_array
```

```python
np.linalg.inv(value_as_array)
```

## Obtener el mayor elemento

```python
value = [ [[21, 83, 42],
           [35, 12, 67]],
          [[56, 98, 24],
           [64, 78, 98]] ]
value_as_array = np.array(value)
```

```python
np.max(value_as_array, axis=0)
```

```python
np.max(value_as_array, axis=1)
```

```python
np.max(value_as_array, axis=2)
```

```python
np.max(value_as_array, axis=-1) #-1: last_possible_axis
```

## Obtener el índice del mayor elemento

```python
value = [ [[21, 83, 42],
           [35, 12, 67]],
          [[56, 98, 24],
           [64, 78, 98]] ]
value_as_array = np.array(value)
```

```python
#56>21, 98>83, 42>24
#64>35, 78>12, 98>67
np.argmax(value_as_array, axis=0)
```

```python
#35>21, 83>12, 67>42
#64>56, 98>78, 98>24
np.argmax(value_as_array, axis=1)
```

```python
#83>42>21, 67>35>12
#98>56>24, 98>78>64
np.argmax(value_as_array, axis=2)
```

```python
np.argmax(value_as_array, axis=-1) #-1: last_possible_axis
```
