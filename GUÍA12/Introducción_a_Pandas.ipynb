También podemos crear nuestro propio DataFrame desde cero:

```python
columns = ["Carrera", "Facultad"]

data = [ ["Ingeniería Informática", "Ciencias e Ingeniería"],
         ["Antropología", "Ciencias Sociales"],
         ["Contabilidad", "Ciencias Contables"],
         ["Física", "Ciencias e Ingeniería"],
         ["Diseño Gráfico", "Arte y Diseño"],
         ["Física", "Ciencias e Ingeniería"] ]

dataset = pd.DataFrame(data=data, columns=columns)

dataset
```

Para agregar/reemplazar filas a un DataFrame haremos uso de la propiedad "loc":

```python
#Agregando filas
dataset.loc[2] = ["No definido", "No definido"]

dataset.loc[6] = ["Contabilidad", "Ciencias Contables"]

dataset
```

```python
#Reemplazando una fila
dataset.loc[1] = ["Derecho", "Derecho"]

dataset
```

Para agregar columnas a nuestro DataFrame haremos uso de los corchetes:

```python
data = [1, 2, "3", 4, "5", 6, "7"]

dataset["Nueva Columna"] = data

dataset
```

Dado que la columna tiene un nombre compuesto por espacios, accederemos a ella mediante los corchetes:

```python
dataset["Nueva Columna"].head(2)
```

Para reemplazar los valores de una columna basta acceder a ella y establecer los nuevos valores:

```python
dataset["Nueva Columna"] = [10, 11, 12, 13, 14, 15, 16]

dataset
```

Si quisieramos reemplazar un valor en valor en específico, haremos uso de la columna y el índice de la fila:

```python
#Haciendo uso de la propiedad "at"
row_index = 5

column_name = "Nueva Columna"

dataset.at[row_index, column_name] = 20

dataset
```

Pandas nos permite manipular fechas mediante el método "Timestamp", que verificará que el dato introducido sea correcto:

```python
#Creamos una fecha cuyo formato es: año mes día hora minuto segundo
date = "20200229013010"

pd.Timestamp(date)
```

```python
#Tratamos de crear la fecha 29 de febrero del 2019 la cual es inválida. Deberíamos ver la impresión "could not convert string to Timestamp"
date = "20190229013010"

try:
  pd.Timestamp(date)
except Exception as e:
 print(e)
```

Reemplazamos el nombre y valores de la columna que creamos con fechas

```python
#Cambiamos el nombre de la columna
dataset.rename(columns={"Nueva Columna": "Timestamp"}, inplace=True)

dataset
```

```python
#Cambiando los valores de la columna. Si ponemos un sólo valor, éste se repetirá para todas las filas
dataset["Timestamp"] = 20

dataset
```

```python
#Cambiando los valores de la columna usando las fechas
date = "20200101"

dataset["Timestamp"] = pd.Timestamp(date)

dataset
```

```python
#Visualizamos el tipo de dato de las columnas
dataset.dtypes
```

En caso ya tengamos las fechas como cadenas de texto, podemos transformar dichas cadenas de texto al tipo de dato fecha:

```python
#Modificamos las fechas a cadenas de texto
dataset["Timestamp"] = "20200101"

dataset
```

```python
#Modificamos las cadenas de texto al tipo de dato fecha
#Para ver los formatos disponibles revisar https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
dataset["Timestamp"] =  pd.to_datetime(dataset["Timestamp"], format="%Y%m%d")

dataset
```

```python
#Visualizamos el tipo de dato de las columnas
dataset.dtypes
```
