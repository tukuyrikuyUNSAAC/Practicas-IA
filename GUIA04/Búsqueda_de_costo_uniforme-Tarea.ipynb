#BÚSQUEDA DE COSTO UNIFORME


La búsqueda de costo uniforme expande el nodo n con el camino de costo más pequeño. Notemos que si todos los costos son iguales, es idéntico a la búsqueda primero en anchura.Los nodos a medida que se expanden se almacenan en una estructura de datos de tipo COLA DE PRIORIDAD(Frontera).
