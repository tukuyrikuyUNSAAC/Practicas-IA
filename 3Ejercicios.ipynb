##EJERCICIO 01

###Enunciado

**Implementar un módulo que permita verificar si una solución de sudoku es o no correcta**

###Modelo

**La cuadrícula de Sudoku consiste en espacios de 9x9.**

- Solo se pueden usar números del 1 al 9.
- Cada bloque de 3×3 solo puede contener números del 1 al 9.
- Cada columna vertical solo puede contener números del 1 al 9.
- Cada fila horizontal solo puede contener números del 1 al 9.

El juego finaliza cuando toda la cuadrícula de Sudoku se completa correctamente con los números.


###Código PYTHON

####Módulo

```python
def verificar_sudoku(sudoku):
  # -- Verificar filas
  vfilas = sum([len({v for v in fila}) for fila in sudoku ]) == 81
  # -- Verificar columnas
  vcolumnas = sum([len({sudoku[f][c] for f in range(9)}) for c in range(9)]) == 81
  # -- Verificar cuadrantes
  coord = [(0,0),(0,3),(0,6),(3,0),(3,3),(3,6),(6,0),(6,3),(6,6)]
  vcuadrantes = sum([len({sudoku[t[0]+f][t[1]+c] for c in range(3) for f in range(3)}) for t in coord]) == 81
  # -- Devolver resultado
  return  vfilas and vcolumnas and vcuadrantes
```

####Programa principal

Definir solución sudoku

```python
# -- Definir solución sudoku
sudoku = [[6,1,8,5,4,7,9,3,2],
          [5,4,2,9,1,3,7,8,6],
          [3,7,9,2,8,6,1,5,4],
          [8,6,7,3,9,4,2,1,5],
          [9,5,1,8,7,2,6,4,3],
          [4,2,3,1,6,5,8,9,7],
          [1,3,4,7,2,8,5,6,9],
          [7,8,6,4,5,9,3,2,1],
          [2,9,5,6,3,1,4,7,8]]
# -- Verificar si es una solución correcta
print('Es correcta' if verificar_sudoku(sudoku) else 'No es correcta')

# -- Definir solución sudoku
sudoku = [[6,1,8,5,4,7,9,3,2],
          [5,4,2,9,1,3,7,8,6],
          [3,7,9,2,8,6,1,5,4],
          [8,6,7,3,9,4,2,1,5],
          [9,5,1,8,7,2,6,4,3],
          [4,2,3,1,6,5,8,9,7],
          [1,3,4,7,2,8,5,6,9],
          [7,8,6,4,5,9,3,2,1],
          [2,9,5,6,3,1,4,8,7]]
# -- Verificar si es una solución correcta
print('Es correcta' if verificar_sudoku(sudoku) else 'No es correcta')
```
