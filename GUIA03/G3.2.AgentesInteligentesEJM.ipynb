Problema de recolección de objetos en un entorno con obstáculos, con un agente reactivo simple.
El agente debe recoger un objeto si hay uno en frente. Si no hay un objeto en frente pero hay un obstáculo, el agente intenta evitar el obstáculo girando a la derecha o izquierda si es posible. Si no puede girar, gira en la dirección contraria a la que estaba yendo originalmente. Si no hay obstáculos ni objetos en frente, el agente simplemente avanza.


```python
import random

# -- Implementación de la clase AgenteReactivoSimple
class AgenteReactivoSimple:
    def __init__(self):
        pass

    def seleccionar_accion(self, percepcion):
        if percepcion["objeto_frente"]:
            return "recoger"
        elif percepcion["obstaculo_frente"]:
            if percepcion["obstaculo_izquierda"]:
                return "derecha"
            elif percepcion["obstaculo_derecha"]:
                return "izquierda"
            else:
                return "derecha" if percepcion["direccion"] == "arriba" else "izquierda"
        else:
            return "adelante"

# ******************** PROGRAMA PRINCIPAL  ********************
# -- Crear objeto AgenteReactivoSimple
agente = AgenteReactivoSimple()

# -- Percibir el entorno
percibir_entorno = input('¿Percibir entorno <S> o <N>?').upper()=='S'

while percibir_entorno:
  # -- Percepción del agente (Asumir que los sensores determinan esta información)
  percepcion = {"objeto_frente": random.choice([True, False]),
                "obstaculo_frente": random.choice([True, False]),
                "obstaculo_izquierda": random.choice([True, False]),
                "obstaculo_derecha": random.choice([True, False]),
                "direccion": random.choice(["abajo", "arriba"])}
  # -- Mostrar percepción (información de los sensores)
  print()
  print('Información de los sensores')
  print('---------------------------')
  print("objeto_frente = ", percepcion["objeto_frente"])
  print("obstaculo_frente", percepcion["obstaculo_frente"])
  print("obstaculo_izquierda", percepcion["obstaculo_izquierda"])
  print("obstaculo_derecha", percepcion["obstaculo_derecha"])
  print("direccion", percepcion["direccion"])
  # -- Procesar la percepción y determinar la acción del agente
  accion_seleccionada = agente.seleccionar_accion(percepcion)
  print("Acción del agente:", accion_seleccionada)
  # -- Nueva percepción del entorno
  percibir_entorno = input('¿Percibir entorno <S> o <N>?').upper()=='S'
```
