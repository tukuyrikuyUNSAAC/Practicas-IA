###Implementación avanzada

```python
# -- Importar las librerías
import random
from pyeasyga.pyeasyga import GeneticAlgorithm
# -- Definir la estructura de los individuos
data = [('pera', 50), ('manzana', 35), ('naranja', 40)]
# -- Crear el objeto "Algoritmo Genético" dando valores a sus hiperparámetros
ga = GeneticAlgorithm(data,
                      population_size=10,
                      generations=20,
                      crossover_probability=0.8,
                      mutation_probability=0.05,
                      elitism=True,
                      maximise_fitness=True)

# -- Definir la función que utilizará el algoritmo genético para "crear individuos"
def crear_individuo(data):
   return [random.randint(0, 1) for _ in range(len(data))]

# -- "Hacer saber" al algoritmo genético que debe utilizar la función definida
ga.create_individual = crear_individuo

# -- Definir la función que utilizará el algoritmo genético para efectuar "el cruce"
def cruce(padre1, padre2):
  crossover_index = random.randrange(1, len(padre1))
  child_1 = padre1[:crossover_index] + padre2[crossover_index:]
  child_2 = padre2[:crossover_index] + padre1[crossover_index:]
  return child_1, child_2

# -- "Hacer saber" al algoritmo genético que debe utilizar la función definida
ga.crossover_function = cruce

# -- Definir la función que utilizará el algoritmo genético para efectuar "la mutación"
def mutacion(individuo):
  mutate_index = random.randrange(len(individuo))
  if individuo[mutate_index] == 0:
    individuo[mutate_index] = 1
  else:
    individuo[mutate_index] = 0

# -- "Hacer saber" al algoritmo genético que debe utilizar la función definida
ga.mutate_function = mutacion

# -- Definir la función que utilizará el algoritmo genético para efectuar "la selección"
def seleccion(poblacion):
  return random.choice(poblacion)

# -- "Hacer saber" al algoritmo genético que debe utilizar la función definida
ga.selection_function = seleccion

# -- Definir la función que utilizará el algoritmo genético para efectuar "la evaluación" de cada individuo
def funcion_aptitud(individuo, data):
  # -- Inicializar puntaje
  puntaje = 0
  # -- Verificar si es un individuo válido
  if individuo.count(1) == 2:
    # -- Calcular puntaje en base a la ganancia de cada fruta
    for (gen,(fruta, ganancia)) in zip(individuo,data):
      if gen == 1:
        puntaje += ganancia
  # -- Devolver resultado
  return puntaje
# -- "Hacer saber" al algoritmo genético que debe utilizar la función definida
ga.fitness_function = funcion_aptitud

# -- Ejecutar el algoritmo genético
ga.run()

# -- Mostrar al "mejor individuo" de "la mejor solución"
print('Mejor individuo')
print(ga.best_individual())

# -- También se puede mostrar todos "los individuos" de "la última generación"
print('\nRelación de individuos de la última generación')
for individual in ga.last_generation():
  print (individual)
```
